Problem Statement 1:Smart Parking Lot Management System

```python
def manage_parking(capacity, logs):

    parked_count = 0

    for action in logs:
        if action == "IN":
            parked_count += 1
        elif action == "OUT":
            parked_count -= 1

    if parked_count >= capacity:
        status = "Full / Over Capacity"
    else:
        status = "Available"

    return parked_count, status


capacity = int(input("Enter the capacity of the parking lot: "))
logs = input("Enter Vehicle Logs (separated by spaces): ").split()

current_parked, parking_status = manage_parking(capacity, logs)
print(f"Currently Parked Vehicles: {current_parked}")
print(f"Parking Status: {parking_status}")
```

Problem Statement 2:Online Food Delivery Time Estimator

```python
def estimate_delivery_time(distance_km, traffic_level, weather):

    base_time = distance_km * 5

    traffic_delays = {"Low": 0, "Medium": 5, "High": 10}

    weather_delays = {"Clear": 0, "Rainy": 5, "Stormy": 10}

    total_time = base_time + traffic_delays.get(traffic_level, 0) + weather_delays.get(weather, 0)
    return total_time

dist = int(input("Enter distance in KM: "))
traffic =input("Enter traffic level (Low, Medium, Hight): ").lower()
weather_cond = input("Enter weather condition (Clean, Rainy, Stormy): ").lower()

eta = estimate_delivery_time(dist, traffic, weather_cond)
print(f"Estimated Delivery Time: {eta} minutes")
```

Problem Statement 3:Movie Theatre Seat Occupancy

```python
def seat_analyzer(total_seats, booked_count):

    occupancy = (booked_count / total_seats) * 100
    if occupancy == 100:
        status = 'Housefull'
    elif occupancy >= 75:
        status = 'Almost Full'
    else:
        status = f'{occupancy:.0f}% Full'
    return f'Occupancy: {occupancy:.0f}%\nShow Status: {status}'


total_seats = int(input("Enter total seats: "))
booked_count = int(input("Enter booked seats: "))
print(seat_analyzer(total_seats, booked_count))
```

Problem Statement 4:Cloud Server Load Classification System

```python
def server_load_classifier(cpu_readings):

    avg_cpu = sum(cpu_readings) / len(cpu_readings)
    if avg_cpu < 50:
        status = 'Normal'
    elif avg_cpu < 80:
        status = 'Warning'
    else:
        status = 'Critical'
    return f'Average CPU Load: {avg_cpu:.0f}\nServer Status: {status}'

cpu_readings=list(map(int, input("Enter CPU readings (space separated): ").split()))
print(server_load_classifier(cpu_readings))

```

Problem Statement 5:Smart Classroom Resource Usage Monitor

```python
def resource_monitor(usage_hours):

    overused = []
    for resource, hours in usage_hours.items():
        if hours > 8:
            overused.append(resource)
    alert = 'Yes' if overused else 'No'
    return f'Overused Resources: {", ".join(overused) if overused else "None"}\nEnergy Alert: {alert}'

usage_hours={"Projector": 6, "AC": 9, "Lights": 4}
print(resource_monitor(usage_hours))

```

Problem Statement 6: Online Event Registration Capacity Controller

```python
def event_registration(capacity, registrations):

    confirmed = min(capacity, registrations)
    waitlisted = max(0, registrations - capacity)
    status = 'Closed' if waitlisted > 0 else 'Open'
    return (f'Confirmed Registrations: {confirmed}\n'
            f'Waitlisted Users: {waitlisted}\n'
            f'Registration Status: {status}')

capacity=int(input("Enter the capacity of the event:"))
registrations=int(input("Enter the number registration: "))
print(event_registration(capacity, registrations))

```
