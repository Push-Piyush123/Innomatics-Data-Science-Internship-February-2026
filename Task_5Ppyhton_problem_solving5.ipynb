```python
# Logic_Building_Task_5:

# Problem 1: Post Engagement Analyzer

def analyze_post_engagement(likes_list):
    """
    Calculates total likes and determines
    whether the post is Viral or Normal.
    """

    total_likes = 0

    # Loop to calculate total likes
    for like in likes_list:
        total_likes += like

    # Check viral condition
    if total_likes >= 1000:
        status = "Viral Post"
    else:
        status = "Normal Engagement"

    return total_likes, status


# Sample Data
likes_data = [200, 300, 250, 300]

total, result = analyze_post_engagement(likes_data)

print("Total Likes:", total)
print("Post Status:", result)
```

```python
# Problem 2: Medicine Stock Alert System

def check_medicine_stock(stock):
    """
    Checks if stock level is sufficient or low.
    """

    if stock < 10:
        return "Low Stock Alert"
    else:
        return "Stock Sufficient"


medicine_stock = 6

print("Medicine Stock:", medicine_stock)
print("Status:", check_medicine_stock(medicine_stock))
```

```python
# Problem 3: Rainfall Adequacy Checker

def rainfall_analysis(rainfall_data, required_level):
    """
    Calculates average rainfall and checks adequacy.
    """

    total_rainfall = 0

    for value in rainfall_data:
        total_rainfall += value

    average = total_rainfall / len(rainfall_data)

    if average >= required_level:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"

    return average, status


season_data = [70, 75, 80, 63]
required = 70

avg, rainfall_status = rainfall_analysis(season_data, required)

print("Average Rainfall:", int(avg))
print("Rainfall Status:", rainfall_status)
```

```python
# Problem 4: Duplicate Account Detection

def detect_duplicate_accounts(username_list):
    """
    Checks if duplicate usernames exist
    using set data structure.
    """

    if len(username_list) != len(set(username_list)):
        return "Yes"
    else:
        return "No"


# Sample Data
usernames = ["john_01", "amy22", "mark77", "john_01"]

print("Duplicate Accounts Found:", detect_duplicate_accounts(usernames))
```

```python
# Problem 5: Appointment Eligibility Checker

def check_appointment_eligibility(age):
    """
    Checks whether patient is eligible
    for appointment based on age.
    """

    if age >= 18:
        return "Eligible"
    else:
        return "Not Eligible"


patient_age = 21

print("Patient Age:", patient_age)
print("Eligibility Status:", check_appointment_eligibility(patient_age))
```

```python
# Problem 6: Premium Crop Price Filter

def filter_premium_crops(price_list):
    """
    Filters crop prices above ₹2000.
    """

    premium_list = []

    for price in price_list:
        if price > 2000:
            premium_list.append(price)

    return premium_list


crop_prices = [1500, 2500, 1800, 3200]

print("Premium Crops:", filter_premium_crops(crop_prices))
```

```python
# Problem 7: Application Health Checker

def check_application_health(error_count):
    """
    Evaluates system health based on error count.
    """

    if error_count == 0:
        return "Healthy"
    elif error_count <= 5:
        return "Minor Issues"
    else:
        return "Critical Issues"


errors = 7

print("Error Count:", errors)
print("System Status:", check_application_health(errors))
```

```python
# Problem 8: Daily Transaction Limit Checker

def validate_transaction(amount):
    """
    Checks if transaction exceeds daily limit ₹50,000.
    """

    daily_limit = 50000

    if amount <= daily_limit:
        return "Approved"
    else:
        return "Rejected"


transaction_amount = 60000

print("Transaction Amount:", transaction_amount)
print("Transaction Status:", validate_transaction(transaction_amount))
```

```python
# Problem 9: Attendance Eligibility System

def check_exam_eligibility(attendance_record):
    """
    Calculates attendance percentage
    and checks exam eligibility.
    """

    present_days = 0

    for status in attendance_record:
        if status == "P":
            present_days += 1

    percentage = (present_days / len(attendance_record)) * 100

    if percentage >= 75:
        eligibility = "Eligible"
    else:
        eligibility = "Not Eligible"

    return percentage, eligibility


attendance_data = ["P", "P", "A", "P", "P"]

percent, exam_status = check_exam_eligibility(attendance_data)

print("Attendance Percentage:", percent)
print("Exam Eligibility:", exam_status)
```

```python
# Problem 10: Smart Electricity Bill Analyzer

def calculate_electricity_bill(units):
    """
    Calculates electricity bill based on slab system
    and classifies usage level.
    """

    bill = 0

    # Slab calculation
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)
    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)

    # Usage classification
    if bill < 500:
        status = "Low Usage"
    elif bill <= 1500:
        status = "Moderate Usage"
    else:
        status = "High Usage"

    return bill, status


units_consumed = 250

total_bill, usage_status = calculate_electricity_bill(units_consumed)

print("Units Consumed:", units_consumed)
print("Total Bill:", total_bill)
print("Usage Level:", usage_status)
```
