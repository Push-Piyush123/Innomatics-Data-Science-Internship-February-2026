### Problem 1: Mobile Recharge Validation System 

```python
"""
    This function checks:
    1. Minimum recharge amount
    2. Whether selected plan is valid
    """
def validate_recharge(amount, valid_plans):
    
    if amount < 50:
        print("Recharge amount must be at least ₹50.")
        return False
    
    if amount not in valid_plans:
        print("Invalid plan selected.")
        return False
    
    return True


valid_plans = [199, 299, 399, 599]

while True:
    try:
        amount = int(input("Enter recharge amount (Available plans: 199, 299, 399, 599): "))
        
        if validate_recharge(amount, valid_plans):
            print("Recharge successful!")
            break  # Exit loop if valid
        
        print("Please try again...\n")
    
    except ValueError:
        print("Please enter a valid numeric amount.\n")
```

### Problem 2: Inventory Reorder Alert System

```python
"""
    This function checks each product's stock.
    If stock is less than 15 → Reorder Alert.
    Otherwise → Stock OK.
    """
def check_inventory(stock_data):
   
    print("\nInventory Status Report:\n")
    
    for product, quantity in stock_data.items():
        if quantity < 15:
            print(f"{product} → Reorder Alert (Stock: {quantity})")
        else:
            print(f"{product} → Stock OK (Stock: {quantity})")


# Taking input from user

inventory = {}

num_products = int(input("Enter number of products to add: "))

for _ in range(num_products):
    product_name = input("Enter product name: ").strip()
    
    # Ensure valid numeric input
    while True:
        try:
            quantity = int(input(f"Enter stock quantity for {product_name}: "))
            if quantity < 0:
                print("Stock cannot be negative. Please enter again.")
                continue
            break
        except ValueError:
            print("Please enter a valid number for quantity.")
    
    inventory[product_name] = quantity

# Call the function
check_inventory(inventory)
```

### Problem 3 — Student Result Processing System

```python
# Calculates average marks and returns Pass/Fail

def process_result(marks_list):
    
    total = 0
    
    # Calculate total using loop
    for mark in marks_list:
        total += mark
    
    average = total / len(marks_list)
    
    print(f"Average Marks: {average:.2f}")
    
    if average >= 50:
        return "Pass"
    else:
        return "Fail"


# Taking input from user
marks = list(map(int, input("Enter student marks separated by space: ").split()))

result = process_result(marks)
print("Result:", result)
```

### Problem 4 — Cab Fare Estimator with Retry Option

```python
"""
    This function checks current system time
    and determines whether it falls under peak hours.
    """

from datetime import datetime, time

def is_peak_hour():
    
    
    # Get current time
    current_time = datetime.now().time()
    
    # Peak hour ranges
    morning_start = time(8, 30)
    morning_end = time(11,00)
    
    evening_start = time(17, 30)
    evening_end = time(20, 0)
    
    # Check if current time falls in the range
    if (morning_start <= current_time <= morning_end) or \
       (evening_start <= current_time <= evening_end):
        return True
    else:
        return False
    
"""
    Base Fare = ₹50
    ₹12 per km
    25% extra during peak hours
    """
def calculate_fare(distance):
   
    base_fare = 50
    fare = base_fare + (distance * 12)
    
    if is_peak_hour():
        print("Peak hour detected — 25% surge pricing applied.")
        fare *= 1.25
    else:
        print("Non-peak hour — normal pricing applied.")
    
    return fare


# Booking Section (enter distance)

while True:
    try:
        distance = float(input("Enter travel distance in km: "))
        
        if distance <= 0:
            print("Distance must be greater than 0.\n")
            continue
        
        total_fare = calculate_fare(distance)
        
        print(f"\nEstimated Fare: ₹{total_fare:.2f}")
        break
    
    except ValueError:
        print("Please enter a valid numeric distance.\n")
```

### Problem 5 — Employee Attendance Eligibility Checker

```python
"""
    Calculates attendance percentage
    and returns eligibility status.
    """
    
def check_attendance(attendance_list):
    
    present_days = 0
    
    # Count number of 'P'
    for status in attendance_list:
        if status.upper() == "P":
            present_days += 1
    
    percentage = (present_days / len(attendance_list)) * 100
    
    print(f"Attendance Percentage: {percentage:.2f}%")
    
    if percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"


# Taking input
attendance = input("Enter attendance record (P/A separated by space): ").split()

eligibility = check_attendance(attendance)
print("Eligibility Status:", eligibility)
```

### Problem 6: Password Strength Checker

```python
"""
    Rules:
    - Minimum length = 8
    - Must contain at least one digit
    - Must contain at least one special character (@#$)
    """

def check_password(password):
    
    if len(password) < 8:
        return "Weak Password (Too Short)"
    
    has_digit = False
    has_special = False
    
    # Check each character using loop
    for ch in password:
        if ch.isdigit():
            has_digit = True
        if ch in "@#$":
            has_special = True
    
    if has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password (Missing digit or special character)"


# Taking input
user_password = input("Enter password: ")

# Call function
strength = check_password(user_password)
print(strength)
```
