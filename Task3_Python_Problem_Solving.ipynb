## Problem Statement 1: Employee Performannce Bonus Eligibility

Given dictionary of employees, find highest performance score employees

```python
employees = {
"Ravi": 92,
"Anita": 88,
"Kiran": 92,
"Suresh": 85
}

highest_score = 0
top_performers = []

# Keep track of highest_score and names 
for name, score in employees.items():
    if score > highest_score:
        highest_score = score
        top_performers = [name]
    elif score == highest_score:
        top_performers.append(name)

# Get the names and display along with score
names = ", ".join(top_performers)
print(f"Top performers Eligible for Bonus: {names} (Score: {highest_score})")
```

## Problem Statement 2: Search Query Keyword Analysis

Given string, count frequency, display keywords with frequency >=2

```python
import string

search_query = "Buy mobile phone buy phone online"

# Converting to lowercase
search_query = search_query.lower()

# Removing punctuation
cleaned_search_query = "".join([i for i in search_query if i not in string.punctuation]).split()

# Counting frequency
frequency = {}
for word in cleaned_search_query:
    frequency[word] = frequency.get(word, 0) + 1

# Use another dictionary and filter frequency of 1 words
res = {word:count for word, count in frequency.items() if count > 1}
print(res)
```

## Problem Statement 3: Sensor Data Validation

Given list, find even readings and display the index along with their readings.

```python
sensor_readings = [3, 4, 7, 8, 10, 12, 5]

res = []

# filter even readings and add them along with their index to res
for i in range(len(sensor_readings)):
    if sensor_readings[i] % 2 == 0:
        res.append((i,sensor_readings[i]))
        
print(res)
```

## Problem Statement 4: Email Domain Usage Analysis

Given list of email IDs, display percentage of each ID user

```python
emails = [
"ravi@gmail.com",
"anita@yahoo.com",
"kiran@gmail.com",
"suresh@gmail.com",
"meena@yahoo.com",
]

# Create an empty dictionary
res = {}

# Frequency of only domain names
for email in emails:
    domain = email.split("@")[1]
    res[domain] = res.get(domain, 0) + 1

# Calculating percentage
for domain, count in res.items():
    count = count/len(emails) * 100
    print(f"{domain}: {count:.0f}%")
```

## Problem Statement 5: Sales Spike Detection

Given, list of sales, detect sales with average above 30%

```python
sales = [1200, 1500, 900, 2200, 1400, 3000]

# Calculate average
avg = sum(sales)/len(sales)


for i in range(len(sales)):
    if sales[i] > avg:
        sale_percent = (sales[i] - avg)/avg * 100    # Calculate % compared to average
        if sale_percent > 30:
            print(f"Day {i+1}: {sales[i]}")
```

## Problem Statement 6 : Duplicate User Id Detection

Given, list of IDs count and display how many times duplicates appear

```python
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

# Empty dictionary
res = {}

# Frequency of users
for i in user_ids:
    res[i] = res.get(i, 0) + 1

# Filter non duplicate users
for user, count in res.items():
    if count > 1:
        print(f"{user} → {count} times")
```
