 Problem Statement 1: Smart Light Controller

 Create a class to control a smart light that can be turned ON or OFF

 and display its status.

```python
class SmartLight:
    
    # Constructor to initialize light name and default status
    def __init__(self, name):
        self.name = name
        self.status = "OFF"
    
    # Method to turn ON the light
    def turn_on(self):
        self.status = "ON"
    
    # Method to turn OFF the light
    def turn_off(self):
        self.status = "OFF"
    
    # Method to display current status
    def display_status(self):
        print(f"{self.name} is {self.status}")


light = SmartLight("Bedroom Light")
light.turn_on()
light.display_status()

```

 Problem Statement 2: Employee ID Card System

 Create a class that stores employee name, ID, and department

 and displays ID card details.

```python
class EmployeeIDCard:
    
    def __init__(self, name, emp_id, department):
        self.name = name
        self.emp_id = emp_id
        self.department = department
    
    def display_id_card(self):
        print("Employee ID Card")
        print("Name:", self.name)
        print("ID:", self.emp_id)
        print("Department:", self.department)


employee = EmployeeIDCard("Rahul", "EMP102", "IT")
employee.display_id_card()

```

 Problem Statement 3: Mobile Contact Record

 Create a class to store contact name and phone number

 and display contact information.

```python
class MobileContact:
    
    def __init__(self, name, phone):
        self.name = name
        self.phone = phone
    
    def save_contact(self):
        print("Contact Saved")
        print("Name:", self.name)
        print("Phone:", self.phone)


contact = MobileContact("Anita", "9876543210")
contact.save_contact()

```

 Problem Statement 4: Product Price Tag Generator

 Create a class that stores product name and price

 and prints a formatted price tag.

```python
class Product:
    
    def __init__(self, name, price):
        self.name = name
        self.price = price
    
    def display_price_tag(self):
        print("Product:", self.name)
        print("Price: ₹" + str(self.price))


product = Product("Headphones", 2499)
product.display_price_tag()

```

 Problem Statement 5: Movie Rating Display System

 Create a class that stores movie name and rating

 and displays movie details.

```python
class Movie:
    
    def __init__(self, name, rating):
        self.name = name
        self.rating = rating
    
    def display_movie_details(self):
        print("Movie:", self.name)
        print("Rating:", self.rating, "/ 5")


movie = Movie("Inception", 4.8)
movie.display_movie_details()

```

 Problem Statement 6: Delivery Address Manager

 Create a class that stores customer name and delivery address

 and prints delivery details.

```python
class DeliveryDetails:
    
    def __init__(self, customer_name, address):
        self.customer_name = customer_name
        self.address = address
    
    def display_delivery_details(self):
        print("Delivery Details")
        print("Customer:", self.customer_name)
        print("Address:", self.address)


delivery = DeliveryDetails("Suman", "Hyderabad")
delivery.display_delivery_details()
```
